```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
import math
import pickle
```

```python
df = pd.read_csv('Real estate.csv')
df.head()
```

```python
X3_distance_to_the_nearest_MRT_station = df['X3 distance to the nearest MRT station']
Y_house_price_of_unit_area = df['Y house price of unit area']
```

```python
#Reshaping the array
Dist_MRT = np.array(X3_distance_to_the_nearest_MRT_station).reshape(-1,1)
Price = np.array(Y_house_price_of_unit_area).reshape(-1,1)
```

```python
#Choosing the model to be Linear Regression
model = LinearRegression()
model.fit(Dist_MRT,Price)
```

```python
Price_prediction = model.predict(Dist_MRT)
```

```python
#Plot
plt.scatter(Dist_MRT,Price, color="blue")
plt.plot(Dist_MRT,Price_prediction, color="red")
plt.title("Linear Regression")
plt.xlabel("Distance to the nearest MRT station")
plt.ylabel("House price of unit area")
```

The plot shows that if travel far from the MRT Statioin, price drops. Which can be observed in the data.

```python
filename = 'linear_regression.pkl'
pickle.dump(model, open(filename, 'wb'))
```
